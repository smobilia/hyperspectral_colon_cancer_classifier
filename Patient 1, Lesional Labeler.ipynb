```python
import os
from os.path import join
import imageio
import numpy as np
import matplotlib.pyplot as plt
```

```python
for (dirname, dirs, files) in os.walk('.'):
    for filename in files:
        oldfile = os.path.join(dirname,filename)
        for i in range(360,381,5): 
            if str(i) in oldfile:
                oldimage=imageio.imread(oldfile)
                newfile = dirname+"__1L__"+str(i)+"nm.tif"
                imageio.imwrite(newfile, oldimage)
        for k in range(390,551,5): 
            if str(k) in oldfile:
                oldimage=imageio.imread(oldfile)
                newfile = dirname+"__1L__"+str(k)+"nm.tif"
                imageio.imwrite(newfile, oldimage)
```
